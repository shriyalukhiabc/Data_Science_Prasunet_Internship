```python
import pandas as pd
data=pd.read_csv('Titanic.csv')
display(data)
```

**Data Cleaning**

```python
data.rename(columns={'SibSp':'siblings/Spouses','Parch':'Parents/Childrens'},inplace=True)
```

```python
data['Embarked']=data['Embarked'].replace('Q','Queenstown')
data['Embarked']=data['Embarked'].replace('S','Southampton')
data['Embarked']=data['Embarked'].replace('C','Cherbourg')
```

```python
data.head()
```

```python
print(sum(data.duplicated()))
print('There is no any duplicate values present in the dataset.')
```

```python
data.isna().sum()
```

```python
data['Cabin'].fillna('not known',inplace=True)
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
```

```python
data['Fare'].fillna(data['Fare'].median(),inplace=True)
```

```python
data.head()
```

```python
data.isna().sum()
```

**Exploratery Data Analysis (EDA)**

**-->Summary Statistics**

```python
print('Average Age:',data['Age'].mean())
print('Median of Age:',data['Age'].median())
print('Mode of Age:\n',data['Age'].mode())
print('Standard deviation of Age:',data['Age'].std())
```

```python
print('Average Fare:',data['Fare'].mean())
print('Median of Fare:',data['Fare'].median())
print('Mode of Fare:\n',data['Fare'].mode())
print('Standard deviation of Fare:',data['Fare'].std())
```

```python
print('Unique values of passenger class:',len(data['Pclass'].unique()))
```

```python
print('Unique values of gender:',len(data['Sex'].unique()))
```

```python
print('Unique values of Embarked:',len(data['Embarked'].unique()))
```

**-->Univariate Analysis**

```python
import matplotlib.pyplot as plt
plt.hist(data['Age'],color='green')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=data['Fare'],color='yellow')
plt.title('Distribution of Fare')
plt.plot()
```

**-->Bivariate Analysis**

```python
survival_rates = data.groupby('Pclass')['Survived'].mean().reset_index()
sns.barplot(x='Pclass', y='Survived', data=survival_rates, palette='viridis')
plt.title('Survival Rate by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

```python
survival_rates = data.groupby('Embarked')['Survived'].mean().reset_index()
sns.barplot(x='Embarked', y='Survived', data=survival_rates)
plt.title('Survival Rate by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

```python
survival_rates = data.groupby('Sex')['Survived'].mean().reset_index()
sns.barplot(x='Sex', y='Survived', data=survival_rates)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

**-->Multivariate Analysis**

```python
survival_rates = data.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()
survival_rates.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Survival Rates by Pclass and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.legend(title='Sex')
```

```python
survival_rates = data.groupby(['Pclass', 'Embarked'])['Survived'].mean().unstack()
survival_rates.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Survival Rates by Pclass and Embarked')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.legend(title='Embarked')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Fare', hue='Survived', palette='viridis', s=100)
plt.title('Scatter Plot of Age vs Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(data, hue='Survived', palette='viridis', diag_kind='kde')
plt.suptitle('Pair Plot of Age, Fare, and Survival Status', y=1.0)
plt.show()
```

**-->Correlation Analysis**

```python
df=data.loc[:,['Survived','Pclass','Age','siblings/Spouses','Parents/Childrens','Fare']]
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
